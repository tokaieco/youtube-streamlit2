```python
#関数化12
def read_excels():

    excel_path= "f_iden_3.xlsm"

    #エクセル読み込み
    due_date_df = pd.read_excel(excel_path, sheet_name="calendar")
    off_day_df = pd.read_excel(excel_path, sheet_name="yukyu")
    skill_df = pd.read_excel(excel_path, sheet_name="skill")
 
 
#     #エラー値を０に変換 エラー防止
    due_date_df.fillna(0,inplace=True) 
    off_day_df.fillna(0,inplace=True)
    skill_df.fillna(0,inplace=True)
#     #データ置換
    off_day_df.replace("◎",2,inplace=True)     # 希望休（◎）を2へ変換
    due_date_df.replace("☆",2,inplace=True)         # 納期（☆）を2へ変換
#品番取り出し
    product_num= due_date_df.loc[:,"品番"]
    
#土日の削除



    return off_day_df, skill_df ,due_date_df, product_num


```
